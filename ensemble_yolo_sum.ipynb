```python
#visdrone ÏÇ¨Ï†ÑÌïôÏäµ Î™®Îç∏ÏóêÏÑúÏùò TEST set ÌèâÍ∞ÄÏΩîÎìú

import torch
import numpy as np
import cv2
import os
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, f1_score, precision_recall_curve

# YOLO Î™®Îç∏ Î°úÎìú
#model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/drive/MyDrive/Colab Notebooks/2024_1/capstone/Model0_visdrone_pre_test/runs/detect/train/weights/best.pt')
model = YOLO('/content/drive/MyDrive/Colab Notebooks/2024_1/capstone/Model0_visdrone_pre_test/runs/detect/train/weights/best.pt')
model.eval()

# Îç∞Ïù¥ÌÑ∞ÏÖã ÌÅ¥ÎûòÏä§ Ï†ïÏùò
class YoloDataset(Dataset):
    def __init__(self, img_dir, label_dir, transform=None):
        self.img_dir = img_dir
        self.label_dir = label_dir
        self.transform = transform
        self.images = os.listdir(img_dir)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.images[idx])
        label_path = os.path.join(self.label_dir, self.images[idx].replace('.jpg', '.txt'))

        image = cv2.imread(img_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        labels = []
        if os.path.exists(label_path):
            with open(label_path, 'r') as file:
                for line in file:
                    class_id, x_center, y_center, width, height = map(float, line.split())
                    labels.append([class_id, x_center, y_center, width, height])

        if self.transform:
            image = self.transform(image)

        return image, np.array(labels)

# Î≥ÄÌôò ÏÑ§Ï†ï Î∞è DataLoader Ï§ÄÎπÑ
transform = transforms.Compose([transforms.ToPILImage(), transforms.Resize((640, 640)), transforms.ToTensor()])
dataset = YoloDataset('/content/drive/MyDrive/Colab Notebooks/2024_1/capstone/drone/rgb-t/each_rgb/test/images', '/content/drive/MyDrive/Colab Notebooks/2024_1/capstone/drone/rgb-t/each_rgb/test/labels', transform)
dataloader = DataLoader(dataset, batch_size=2, shuffle=False)
```

```python
from ultralytics import YOLO  # YOLOv8 Î™®Îç∏ ÌÅ¥ÎûòÏä§
import torch
from pathlib import Path
```

```python
# Í∞ÄÏ§ëÏπò ÌååÏùº Í≤ΩÎ°ú
weights_rgb = '/content/drive/MyDrive/Colab Notebooks/2024_1/capstone/drone/rgb-t/model/runs/detect/train/weights/best.pt'
weights_thermal = '/content/drive/MyDrive/Colab Notebooks/2024_1/capstone/drone/rgb-t/model_thermal/runs/detect/train/weights/best.pt'

# Î™®Îç∏ Ï¥àÍ∏∞Ìôî Î∞è Í∞ÄÏ§ëÏπò Î°úÎìú
model_rgb = YOLO(weights_rgb)
model_thermal = YOLO(weights_thermal)
```

```python
# Ïù¥ÎØ∏ÏßÄ Ìè¥Îçî Í≤ΩÎ°ú ÏÑ§Ï†ï
rgb_images_path = Path('/content/drive/MyDrive/Colab Notebooks/2024_1/capstone/drone/rgb-t/images_origin/rgb/val')
thermal_images_path = Path('/content/drive/MyDrive/Colab Notebooks/2024_1/capstone/drone/rgb-t/images_origin/thermal/val')

# RGB Ïù¥ÎØ∏ÏßÄ ÏòàÏ∏° ÏàòÌñâ
results_rgb = [model_rgb(str(img_path)) for img_path in rgb_images_path.glob('*.jpg')]

# Ï†ÅÏô∏ÏÑ† Ïù¥ÎØ∏ÏßÄ ÏòàÏ∏° ÏàòÌñâ
results_thermal = [model_thermal(str(img_path)) for img_path in thermal_images_path.glob('*.jpg')]
```

```python
# ÏòàÏ∏° Í≤∞Í≥º ÌÜµÌï©
combined_preds = torch.cat([result.pred[0] for result in results_rgb + results_thermal], 0)

# NMS Ï†ÅÏö©
conf_thres = 0.25  # Ïã†Î¢∞ÎèÑ ÏûÑÍ≥ÑÍ∞í
iou_thres = 0.45   # IOU ÏûÑÍ≥ÑÍ∞í
nms_indices = torch.ops.torchvision.nms(combined_preds[:, :4], combined_preds[:, 4], iou_thres)

# NMS ÌõÑÏùò Î∞îÏö¥Îî© Î∞ïÏä§ Ï∂îÏ∂ú
final_boxes = combined_preds[nms_indices]
```

```python
# ÏµúÏ¢Ö Î∞îÏö¥Îî© Î∞ïÏä§ Ï∂úÎ†•
print(f'Final bounding boxes after NMS: {final_boxes}')
```
